**Q1.  입력 받은 년도가 윤년인지 아닌지 판단하는 프로그램을 구현하세요. 윤년은 다음과 같은 규칙을 따릅니다.**

- 윤년은 연도가 윤년은 연도가 4의 배수이면서, 100의 배수가 아닐 때 또는 400의 배수일 때입니다.
예를 들어, 2012년은 4의 배수이면서 100의 배수가 아니라 윤년입니다. 1900년은 100의 배수이고 400의 배수는 아니기 때문에 윤년이 아닙니다. 하지만, 2000년은 400의 배수이기 때문에 윤년입니다.

```python
year = int(input())

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            is_leap_year = True
        else:
            is_leap_year = False
    else:
        is_leap_year = True
else:
    is_leap_year = False

if is_leap_year:
    print(f'{year} is a leap year')
    
else:
    print(f'{year} is not a leap year')
```

**Q3. 다음 조건을 만족하는 함수를 define 하세요.**

- 함수 이름은 maxAvg()이며, 사용자가 입력한 실수들의 최댓값과 평균을 반환합니다.
- 사용자는 2개 이상 5개 이하의 실수들을 입력합니다.

```python
def maxAvg() :
    print("멈추고 싶으면 음수를 입력하세요")
    lst = []
    
    try :

        while True : 
            a = float(input("숫자를 입력하십시오:"))
            
            if a < 0 :
                if len(lst) < 2:
                    raise Exception("숫자를 2개 이상 입력하세요\n")
                elif len(lst) > 5:
                    raise Exception("숫자를 5개 이하로 입력하세요\n")
                else : 
                    break
                    
            else :
                lst.append(a)
        
        print("입력한 숫자들의 최댓값:", max(lst))
        print("입력한 숫자들의 최댓값:", sum(lst)/len(lst))
            
    except Exception as e:

        print("예외가 발생했습니다.", e)
    
maxAvg()
```

**Q5. 정렬 알고리즘을 사용하여 지원자들의 번호대로 이름을 출력하는 프로그램을 만드세요.**

면접관 나혁이는 CUAI 면접 순서에 맞게 지원자들의 줄을 세우려고 합니다. 이미 부여된 면접 번호대로 사람들의 줄을 세울 예정입니다. 면접 순서가 적힌 쪽지에는 다음과 같이 적혀 있습니다.

- 1번 Jerry, 2번 Mike, 3번 Tom, 4번 Rachel, 5번 Jackson

정렬 알고리즘을 사용하여 지원자들의 번호대로 이름을 출력하는 프로그램을 만드세요. 첫째 줄에는 지원자들의 수 N, 둘째 줄부터는 지원자들의 이름과 면접 순서가 입력됩니다.

```python
interview = {}
idx = int(input("입력: "))

for i in range(idx) :
    a, b = input().split(" ")
    interview[int(b)] = a

interview = sorted(interview.items())

for value in interview :
    print(value[1], end = " ")

    
```
